```python
from lxml import etree
```

```python
infile = 'input/Centennial_invoices_20211027.xml'
outfile = 'output/Centennial_invoices_fixed_20211028.xml'
```

```python
# Specify the XMLParser with recover option to catch any '&' that 
# have not been properly escaped in the source file.
tree = etree.parse(infile, parser=etree.XMLParser(recover=True))
funds = tree.xpath('//fundID')
print(len(funds))
```

```python
for fund in funds:
    print('original: ' + fund.text)
    fund_original = fund.text
    
    # find the next sibling element with tag 'fiscalCycle'
    fiscal = fund.xpath('../fiscalCycle')
    # check if one such sibling was found
    if len(fiscal) > 0:
        # replace value of the fundID element with the fundID concatenated with the fiscal year
        fund.text = fund_original + '-' + fiscal[0].text
    else:
        # there are cases where there is no fiscalCycle sibling. Try to infer the fiscal year from another
        # fiscalCycle element elsewhere within the same invoice
        print('WARNING: no fiscal cycle found for fund ID ' + fund_original + ' on line ' + str(fund.sourceline))
        parent_invoice = fund.xpath('../..//fiscalCycle')
        if len(parent_invoice) > 0:
            inferred_year = parent_invoice[0].text
            print('Using other fiscalCycle')
        else:
            fiscal_created = fund.xpath('../..//createdInFiscalCycle')
            if len(fiscal_created) > 0:
                inferred_year = fiscal_created[0].text
                print('Using createdInFiscalCycle')
            else:
                print('No fiscal year found')
        fund.text = fund_original + '-' + inferred_year
    print('modified: ' + fund.text)
tree.write(outfile, pretty_print=True)
```
